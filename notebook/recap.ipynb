```python
import matplotlib.pyplot as plt
import numpy as np

# Data F1 per skenario (urut A..L)
labels = list("ABCDEFGHIJKL")
f1 = [0.8990, 0.9068, 0.9294, 0.1103, 0.1103, 0.9082,
      0.7300, 0.7052, 0.7425, 0.1900, 0.3051, 0.6124]

# Warna: gunakan colormap 'magma' agar tidak mainstream & konsisten
cmap = plt.get_cmap("magma")
colors = [cmap(i / (len(f1) - 1)) for i in range(len(f1))]

plt.figure(figsize=(10, 4.5))
bars = plt.bar(labels, f1, color=colors, edgecolor="black", linewidth=0.6)
plt.title("F1-Score Comparison Between Scenarios (Aâ€“L)")
plt.xlabel("Skenario")
plt.ylabel("F1-Score")
plt.ylim(0, 1.0)
for b, v in zip(bars, f1):
    plt.text(b.get_x() + b.get_width()/2, v + 0.02, f"{v:.2f}",
             ha="center", va="bottom", fontsize=9)
plt.tight_layout()
plt.show()
```
